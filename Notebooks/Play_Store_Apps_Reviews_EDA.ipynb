## Cleaning Data ##

<h4><strong>Check Null / Mising Data</strong></h4>

```python
# -- View Null data --
null = original_data.isna().sum() / original_data.shape[0] * 100 # Percent of null data per variable

sns.barplot(x = null[null.values>0.7].index, y = null[null.values>0.7].values) # Filter > 0.7% of missing data

# -- Graph --
plt.xlabel('')
plt.ylabel('Percent (%)')
plt.xticks(rotation=90)
plt.title('Missing data per variable')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

After analyzing the missing data, particularly in the developer's website column, I have decided, following recommendations, not to perform imputations or eliminations. Instead, I will choose to categorize the applications according to the presence or absence of this data as a Boolean variable.

**Clean Data**

```python
df = original_data.copy() # df could be modified

# Columns to be deleted
df.drop(columns = ['App Id', 'Maximum Installs', 'Currency',
                   'Size', 'Developer Email', 'Developer Id', 'Released', 'Last Updated',
                   'Editors Choice', 'Scraped Time','Minimum Android'],
        inplace=True)
```

```python
# -- Transfor binary categorical data in 0 and 1 --

# -- Columns with original Bool data --
df['Free'] = df['Free'].astype(int)
df['Ad Supported'] = df['Ad Supported'].astype(int)
df['In App Purchases'] = df['In App Purchases'].astype(int)

# -- Apps with missing values to Bool value --
# {0 -> Data absence (False), 1 -> Data precense (True)}
df['Developer Website'] = df['Developer Website'].notna().astype(int)
df['Privacy Policy'] = df['Privacy Policy'].notna().astype(int)
```

In general, the data are presented in a clean state, with few imperfections. The only manipulations carried out were Boolean variable conversions and the exclusive selection of columns relevant to our analysis.

```python
# -- No rated apps --

rate_0 = df[df['Rating'] == 0] # Filter by Ratin equals 0 stars on reviews
rate_0_count = rate_0.shape[0] # Number of Apps with 0 stars

no_imputed = df.copy()         # Copy of the non imputed dataframe

print(f'\nNo Rated Apps: {rate_0_count / df.shape[0] * 100 :.2f}%\nRated Apps : {(1-rate_0_count/ df.shape[0])* 100 :.2f}%')

df = df[df['Rating'] != 0]  # Delete data wit Rating equals 0
```

After this observation, we note that approximately 46% of the data have a rating of 0 stars, while the minimum is 1. Since the central objective of the project is to predict the rating of an application based on its characteristics, we will choose to impute this missing data. It is important to note that these 0-star rated applications also lack user reviews.

### **Q3)** How Category is correlated to other features?

```python
cat_count = df['Category'].value_counts() # Count each app by the differents 26 categroies

# -- Graph count of apps by category --

plt.figure(figsize=(15, 7))
sns.barplot(x = cat_count.index, y = cat_count.values, palette = sns.color_palette("Set2"))
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Apps by category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=80)
plt.show()
```

In this distribution, it is evident to observe how a considerable part of the applications on the market are mainly grouped in categories such as Education, Tools, Entertainment and Music & Audio. In contrast, a smaller proportion of the supply is found in categories such as Music, Libraries and Demos, Parenting and Comics.

```python
cat_download = df.groupby('Category')['Minimum Installs'].agg(['sum', 'mean']) # Sum total installs by category sum and mean
cat_download = cat_download.sort_values(by=['sum'], ascending=False)

# -- Graph total Downloads by category --
plt.figure(figsize=(15, 7))
sns.barplot(x = cat_download.index, y = cat_download['sum'], palette = sns.color_palette("Set2"))
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Downloads by category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=80)
plt.show()
```

In the distribution of the number of applications versus the number of downloads, we observe that the "Tools" category continues to maintain its dominant position in the comparison. However, there is a significant increase in the "Communication" category, which moves up 17 positions in relation to the number of applications versus downloads. Similarly, the "Productivity" category also shows an increase of 7 positions in this comparison. On the other hand, it is important to mention that the "Medical" category experiences a notable decline, dropping 19 positions in the comparison between the number of applications and downloads. These changes reflect fluctuations in user preferences and demands in the field of mobile applications.

```python
cat = pd.concat([cat_download, cat_count], axis=1, join='outer')  # concant the other df
cat['saturation'] = (cat['sum']/cat['count'])                     # Create the "saturation index"
cat.sort_values(by='saturation', ascending=False, inplace = True) # Sort values

# -- Graph --
plt.figure(figsize=(15, 7))
sns.barplot(x = cat.index, y = cat.saturation, palette = sns.color_palette("Set2"))
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('saturation index')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=80)
plt.show()
```

In our exploratory data analysis (EDA), we have introduced a fundamental indicator to understand the dynamics between supply and demand in the mobile application market: the saturation index. This index is calculated by dividing the number of downloads by the number of apps per category. A high index indicates good demand relative to the number of applications available in that category. Conversely, a lower index suggests that the market is saturated with applications with a relatively lower download rate. This indicator provides valuable information on the competitiveness and viability of the different categories of applications in the mobile market.

### **Q5)** What decisions did the development team make and how do they relate to the project?

```python
# -- Graph Developer website presence with rating --
sns.boxplot(data = df, x='Developer Website', y='Rating')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks([1, 0], ['True', 'False'])
plt.title('Developer website info')
plt.show()

# -- Print information about absence data --
website_info = df[df['Developer Website'] == 0].shape[0]
print(f'\nPercentage of data with no website information: { website_info / df.shape[0] * 100 :.2f}%')
```

The last observation indicates that, despite the initial belief that the inclusion of the developer's website might have an impact on the application's rating, this does not appear to be the case based on the analysis

The same is observed with respect to the privacy policy. Despite the initial assumption that its presence or absence could influence the application's rating, the analysis suggests that this is not necessarily true.

```python
sns.boxplot(data = df, x='Ad Supported', y='Rating')

plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks([1, 0], ['True', 'False'])

plt.title('Ad Supported')
plt.show()

add = df[df['Ad Supported'] == 0].shape[0]
print(f'\nNoAdds in App: {add / df.shape[0] * 100 :.2f}%\nAdds in App: {(1-add / df.shape[0]) * 100 :.2f}%')
```

```python
plt.figure(figsize=(15, 7))
sns.boxplot(data = df, x='Content Rating', y='Rating', palette = sns.color_palette("Set2"))

plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=80)
plt.title('Content rating')
plt.show()
```

The content rating does not appear to have a significant impact on the application rating.

```python
total_by_content = df['Content Rating'].value_counts()
content_group = df.groupby('Content Rating')[['Developer Website', 'Ad Supported','In App Purchases', 'Privacy Policy', 'Free']].sum()

sns.heatmap(content_group.div(total_by_content, axis=0).T,  cmap='hot', annot=True, fmt=".2f")

plt.title('Heatmap de contenido')
plt.ylabel('Some características')
plt.xticks(rotation=80)

plt.show()
```


In this last graph, we observe the relationship between an app's rating and variables considered as decisions of the development team. We confirm that unrated apps tend to have a lower rate of developer information and privacy policies, while having a higher number of ads but fewer in-app purchases. On the other hand, apps with +18 ratings tend to have fewer purchases but more ads, and all of these apps are free. Interestingly, despite being a less frequent indicator, the apps with a higher rate of purchases are those rated +10, which tend to be aimed at the accessible video game market for children and also have the highest rate of in-app ads.
